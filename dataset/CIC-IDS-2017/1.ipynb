```python
import pandas as pd

paths = [
    "/root/autodl-fs/processed_data/CIC-IDS-2017-session-srcIP_srcPort_dstIP_dstPort_proto/Thursday-WorkingHours/Thursday-WorkingHours-flow.csv",
    "/root/autodl-fs/processed_data1/CIC-IDS-2017-session-srcIP_srcPort_dstIP_dstPort_proto/Thursday-WorkingHours/Thursday-WorkingHours-flow.csv"
]

for i, path in enumerate(paths, 1):
    # 分块计数，节省内存
    row_count = 0
    for chunk in pd.read_csv(path, chunksize=100000):
        row_count += len(chunk)
    print(f"文件{i}行数: {row_count}")
```

```python
import pandas as pd

# 文件路径
file_path = "/root/autodl-fs/processed_data/CIC-IDS-2017-session-srcIP_srcPort_dstIP_dstPort_proto/Thursday-WorkingHours/Thursday-WorkingHours-flow.csv"

# 只读取前几行和必要的列
df_head = pd.read_csv(file_path, nrows=5)

print("前5行前12列数据:")
print(df_head.iloc[:, :12])

print("\n" + "="*50)

# 快速检查是否排序（只检查前1000行）
df_sample = pd.read_csv(file_path, nrows=1000, usecols=['conn.ts'] if 'conn.ts' in df_head.columns else [])

if 'conn.ts' in df_sample.columns:
    # 检查前1000行是否排序
    is_sorted = (df_sample['conn.ts'].diff().dropna() >= 0).all()
    print(f"\n检查前1000行是否按 'conn.ts' 升序: {'是' if is_sorted else '否'}")
else:
    print(f"\n文件没有 'conn.ts' 列")
    print(f"前几行列名: {df_head.columns[:10].tolist()}")
```
